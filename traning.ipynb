```python
import pandas as pd

df = pd.read_csv("C:\\Users\\Swapnil\\Downloads\\insurance.csv")
df.head()
```

```python
df.describe()
```

```python
df['BMI'] = round(df['Weight'] / ((df['Height']*0.01)**2),2)
```

```python
df.head()
```

```python
x = df.drop(['PremiumPrice'],axis=1)
y = df['PremiumPrice']
```

```python
from sklearn.preprocessing import  StandardScaler
std = StandardScaler()
X = std.fit_transform(x)
x_scaled = pd.DataFrame(X,columns=x.columns)
```

```python
from sklearn.linear_model import Ridge,Lasso,LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import accuracy_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

```python
x_train
```

```python
# Create a random forest regresson
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

rf = RandomForestRegressor(n_estimators=100, random_state=42)

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

grid_search = GridSearchCV(estimator=rf, param_grid=param_grid,
                           cv=3, n_jobs=-1, verbose=2, scoring='r2')
```

```python
grid_search.fit(x_train, y_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
```

```python
best_rf = grid_search.best_estimator_
rf_pred = best_rf.predict(x_test)
```

```python
mse = mean_squared_error(y_test, rf_pred)
mae = np.mean(np.abs(y_test - rf_pred))
r2 = r2_score(y_test, rf_pred)

print(f"Mean Absolute Error {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```
